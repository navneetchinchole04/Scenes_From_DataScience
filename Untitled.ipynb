```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, urlparse
from collections import deque

# Starting URL
start_url = 'https://example.com'

# Max pages to visit
max_pages = 10

# Queues and sets to track progress
queue = deque([start_url])
visited = set([start_url])

print("ğŸš€ Starting crawl...\n")

while queue and len(visited) < max_pages:
    current_url = queue.popleft()
    print(f"ğŸŒ Crawling: {current_url}")

    try:
        response = requests.get(current_url, timeout=10)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"âŒ Error: {e}")
        continue

    soup = BeautifulSoup(response.content, 'lxml')

    # Find all links on the current page
    for link in soup.find_all('a', href=True):
        href = link.get('href')
        full_url = urljoin(current_url, href)

        # Filter only same-domain links (optional)
        if urlparse(full_url).netloc == urlparse(start_url).netloc:
            if full_url not in visited:
                visited.add(full_url)
                queue.append(full_url)

print("\nâœ… Crawl finished.")
print(f"ğŸ”— Total unique links found: {len(visited)}\n")

for url in visited:
    print(url)
```
